```python
def read_file(fn):
    with open(fn, "rb") as f:
        return f.read()
words = read_file("1000_dict.txt").split()
for i in range(len(words)):
    words[i]=words[i].decode()
    
valid_range=[32]+[x for x in range(97,123)]
words=[" "]+words
cipher1 = list(read_file("c1.bin"))
cipher2 = list(read_file("c2.bin"))
xor=[cipher1[i]^cipher2[i] for i in range(len(cipher1))]


reverse={}
dict={}


str="abcdefghijklmnopqrstuvwxyz"
for i in range(26):
    dict[i+97]=str[i]
dict[32]=" "       
for key in dict:
    reverse[dict[key]]=key

def check_XOR(array1, array2):
    for i in range(len(array1)):
        if array1[i]^array2[i]!=xor[i]:
            return False
    return True

def num_letter_convert(num):
    #converts one number to letter
    return dict[num]


def letter_num_convert(letter):
    #converts one letter to number
    return reverse[letter]

def split_space(array, s=32):
    #splits list into sublists which do not contain 32
    s=32
    if s not in array:
        return [array]
    while s in array:
        i=array.index(s)
        prefix=array[:i]
        suffix=array[i+1:]
        return split_space(prefix)+split_space(suffix)

def list_to_word(array):
    total=""
    for i in array:
        total+=num_letter_convert(i)
    return total

def many_list_to_words(array):
    words=[]
    for subarray in array:
        words.append(list_to_word(subarray))
    return words

def check_word(word, exact=False):
    if exact: 
        return (word in words)
    if not exact:

        for true_word in words:
            if true_word[:len(word)]==word:
                return True
        return False

def check_array_of_words(array):   
    for i in range(len(array)):
        word=array[i]
        if i<len(array)-1:
            if not check_word(word, True):
                return False
        if i==len(array)-1:
            if not check_word(word):
                return False
    return True

def total_check(array1, array2):
    if not check_XOR(array1, array2):
        return False
    X=check_array_of_words(many_list_to_words(split_space(array1))) 
    Y=check_array_of_words(many_list_to_words(split_space(array2))) 
    return (X and Y)
```
